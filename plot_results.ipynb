```python
import json
import matplotlib.pyplot as plt
import numpy as np

results = []
with open('results.jsonl') as f:
    for line in f:
        results.append(json.loads(line))

p = [r['noise_strength'] for r in results]
throughput = [r['throughput'] for r in results]
psr = [r['psr'] for r in results]
fidelity = [r['fidelity'] for r in results]
n_errors = [r['n_errors'] for r in results]
n_kept = [r['n_kept'] for r in results]

print(f"{'p':<8} {'Shots/s':>10} {'PSR':>8} {'Errors':>8} {'Kept':>10} {'Fidelity':>12}")
for r in results:
    print(f"{r['noise_strength']:<8} {r['throughput']:>10.0f} {r['psr']:>8.4f} {r['n_errors']:>8d} {r['n_kept']:>10d} {r['fidelity']:>12.8f}")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4), dpi=1000)

# Throughput
axes[0].plot(p, [t/1e6 for t in throughput], 'o-', color='tab:olive')
axes[0].set_xlabel('Noise strength p')
axes[0].set_ylabel('Throughput (M shots/s)')
axes[0].set_title('Sampling Throughput')
axes[0].grid(True, alpha=0.3)

# PSR
axes[1].plot(p, psr, 's-', color='tab:purple')
axes[1].set_xlabel('Noise strength (circuit-level) p')
axes[1].set_ylabel('Post-selection rate')
axes[1].set_title('Post-Selection Rate')
axes[1].grid(True, alpha=0.3)

# Fidelity
ler = np.array([r['n_errors'] / r['n_kept'] for r in results])
ler_err = np.array([np.sqrt(r['n_errors']) / r['n_kept'] for r in results])
axes[2].errorbar(p, ler, yerr=ler_err, fmt='d-', color='tab:green', 
                 capsize=3, label=r"accel. sim. data")
# axes[2].plot(p, 1-np.array(fidelity), 'd-', color='tab:cyan', label = r"accel. sim. data")
axes[2].set_xlabel('Noise strength p')
axes[2].set_ylabel('LER')
axes[2].set_title(r'$d=3$ cultivation (degenerate) noiseless projection')
axes[2].grid(True, alpha=0.3)
axes[2].set_yscale("log")
axes[2].set_xscale("log")
pp = np.arange(1.5e-4,6e-3,1e-3)
axes[2].plot(pp, 800*pp**3, '--',color='black',label = r"LER $= 800p^{3}$")
axes[2].legend()

plt.tight_layout()
plt.savefig('benchmark.png', dpi=150, bbox_inches='tight')
plt.show()
```
